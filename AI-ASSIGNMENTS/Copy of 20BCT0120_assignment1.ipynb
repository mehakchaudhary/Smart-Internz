
Task - 1
 Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)

```python
import pandas as pd
import numpy as np


np.random.seed(50)
data = np.random.randn(4, 4)
dF = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
print(dF)
```

Task - 2 Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4'

```python
dF.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']
print(dF)
```

Task - 3 Find the descriptive statistics of the 'df' dataframe.

```python
statistics = dF.describe()
print(statistics)
```

Task - 4 Check for the null values in 'df' and find the data type of the columns.

```python
null_values = dF.isnull().sum()
data_types = dF.dtypes
print("Null Values:")
print(null_values)
print("\nData Types:")
print(data_types)
```

Task - 5 Display the 'Random value 2' & 'Random value 3' columns with location method and index location method.

```python
location_columns = dF[['Random value 2', 'Random value 3']]
print("Location Method:")
print(location_columns)

# Display columns using the index location method
index_location_columns = dF.iloc[:, [1, 2]]
print("\nIndex Location Method:")
print(index_location_columns)
```
